**Task 2: Multi-Input Prompt System (15%)**


```python
# using multiple inputs like topic, audience and tone
multi_input_prompt = PromptTemplate(
    input_variables=["topic_name", "audience_type", "tone_type"],
    template="Explain {topic_name} for {audience_type} in a {tone_type} tone"
)

# testing with different values
print(multi_input_prompt.format(topic_name="AI", audience_type="beginners", tone_type="friendly"))
print(multi_input_prompt.format(topic_name="Python", audience_type="kids", tone_type="fun"))
print(multi_input_prompt.format(topic_name="Deep Learning", audience_type="engineers", tone_type="technical"))
```

**Task 3: Prompt Variations Engine (15%)**

```python
# creating different styles of prompts

# teaching style
teaching_prompt = PromptTemplate(
    input_variables=["topic_name"],
    template="Explain {topic_name} step by step in a clear manner"
)

# interview style
interview_prompt = PromptTemplate(
    input_variables=["topic_name"],
    template="Generate 3 interview questions related to {topic_name}"
)

# storytelling style
story_prompt = PromptTemplate(
    input_variables=["topic_name"],
    template="Explain {topic_name} using a simple story example"
)

# testing all
test_topic = "Machine Learning"

print(teaching_prompt.format(topic_name=test_topic))
print(interview_prompt.format(topic_name=test_topic))
print(story_prompt.format(topic_name=test_topic))
```

**Task 4: ChatPromptTemplate System (15%)**

```python
# function to create chat prompt based on role
def create_chat_prompt(role_type, topic_name):

    # defining system behavior
    if role_type == "teacher":
        system_msg = "You are a teacher who explains concepts in a simple way."
    elif role_type == "interviewer":
        system_msg = "You are an interviewer asking logical questions."
    elif role_type == "motivator":
        system_msg = "You are a motivational speaker who inspires people."
    else:
        system_msg = "You are a helpful assistant."

    # creating chat template
    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic_name}")
    ])

    return chat_prompt.format_messages(topic_name=topic_name)

# testing
print(create_chat_prompt("teacher", "Neural Networks"))
```

**Task 5: Input Validation Layer (10%)**

```python
# function to validate inputs
def check_inputs(audience_type, tone_type):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # checking audience
    if audience_type not in valid_audience:
        print("Invalid audience detected, using default = beginner")
        audience_type = "beginner"

    # checking tone
    if tone_type not in valid_tone:
        print("Invalid tone detected, using default = formal")
        tone_type = "formal"

    return audience_type, tone_type

# test
print(check_inputs("kids", "funny"))
```

**Task 6: Prompt Generator App (15%)**

```python
# Task 6

# main function to generate prompt dynamically
def generate_prompt(topic_name, audience_type, tone_type, style_type):

    # first validate inputs
    audience_type, tone_type = check_inputs(audience_type, tone_type)

    # choosing template based on style
    if style_type.lower() == "teaching":
        prompt_obj = PromptTemplate(
            input_variables=["topic_name", "audience_type", "tone_type"],
            template="Explain {topic_name} for {audience_type} in a {tone_type} teaching style"
        )

    elif style_type.lower() == "interview":
        prompt_obj = PromptTemplate(
            input_variables=["topic_name", "audience_type", "tone_type"],
            template="Ask interview questions on {topic_name} for {audience_type} in a {tone_type} tone"
        )

    elif style_type.lower() == "storytelling":
        prompt_obj = PromptTemplate(
            input_variables=["topic_name", "audience_type", "tone_type"],
            template="Explain {topic_name} for {audience_type} in a {tone_type} storytelling style"
        )

    else:
        return "Invalid style selected"

    return prompt_obj.format(topic_name=topic_name, audience_type=audience_type, tone_type=tone_type)

# test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7: Template Reusability Test (10%)**

```python
# using one template multiple times
reusable_prompt = PromptTemplate(
    input_variables=["topic_name", "audience_type", "tone_type"],
    template="Explain {topic_name} for {audience_type} in a {tone_type} tone"
)

# different inputs
sample_inputs = [
    ("AI", "beginner", "formal"),
    ("Python", "intermediate", "casual"),
    ("Data Science", "expert", "formal"),
    ("Machine Learning", "beginner", "fun"),
    ("Cloud Computing", "expert", "casual")
]

# running loop
for t, a, tone in sample_inputs:
    print(reusable_prompt.format(topic_name=t, audience_type=a, tone_type=tone))
```
